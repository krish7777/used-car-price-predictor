```python
from sklearn.model_selection import train_test_split
```

```python

x_train, x_test, y_train, y_test = train_test_split(df.loc[:, df.columns != 'Price'],df[['Price']],
                                                    test_size=0.3,
                                                    random_state=0
                                                    )
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
y_pred_mult_lin_reg = regressor.predict(x_test)
y_pred_mult_lin_reg
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test, y_pred_mult_lin_reg))
print(metrics.mean_squared_error(y_test, y_pred_mult_lin_reg))
print(metrics.r2_score(y_test,y_pred_mult_lin_reg))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred_mult_lin_reg)))
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
```

```python
new_x_test = sc_X.fit_transform(x_test)
new_y_test = sc_y.fit_transform(y_test)
new_x_test
```

```python
from sklearn.svm import SVR
```

```python
regressor = SVR(kernel = 'rbf')
regressor.fit(new_x_test, new_y_test)
```

```python
y_pred_svr=sc_y.inverse_transform(regressor.predict(x_test))
```

```python
y_pred_svr
```

```python
print(metrics.mean_absolute_error(y_test,y_pred_svr))
print(metrics.mean_squared_error(y_test, y_pred_svr))
print(metrics.r2_score(y_test,y_pred_svr))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred_svr)))
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred_des_tree=regressor.predict(x_test)
```
